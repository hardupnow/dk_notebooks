This example will help you to to set up a geographical view on cylindrical projection. 
You can change the view playing with the values for *subpage_lower_left_latitude*,  *subpage_lower_left_longitude*, *subpage_upper_right_latitude* and *subpage_upper_right_longitude*

```python
from Magics import macro as magics
```

```python

#Setting the coordinates of the geographical area
projection = magics.mmap(subpage_map_projection = 'cylindrical',
		  subpage_lower_left_latitude= 40.,
		  subpage_lower_left_longitude= -10.,
		  subpage_upper_right_latitude= 60.,
		  subpage_upper_right_longitude= 10)

#Coastlines setting
coast = magics.mcoast( map_grid =  "on",
                map_grid_colour  =  "tan",
                map_coastline_land_shade  =  'on',
                map_coastline_land_shade_colour  =  'cream',
                map_coastline_colour =  "tan",
                map_coastline_resolution = "high",
		)

#Title settings
title = magics.mtext(
	  text_lines = ["<font size='1'>High resolution coastline</font>"],
	  text_justification = "left",
	  text_font_size = 0.8,
	  text_colour = "charcoal")


magics.plot(projection, coast, title)
```

The coastlines object [*mcoast*](https://confluence.ecmwf.int/display/MAGP/Coastlines) comes with a long list of paremeters to customise the visualisation. 

```python
# Grey coastlines
coast = magics.mcoast( map_grid =  "on",
                map_grid_colour  =  "charcoal",
                map_coastline_land_shade  =  'on',
                map_coastline_land_shade_colour  =  'grey',
                map_coastline_colour =  "charcoal",
                map_coastline_resolution = "high",
		)

magics.plot(projection, coast)
```
